```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel('heightWeight.xlsx')
```

```python
df
```

```python
plt.scatter( df['Weight (kg)'], df['Height (cm)'])
plt.xlabel('Weight')
plt.ylabel('Height')
```

```python
df.corr()
```

```python
import seaborn as sns
sns.pairplot(df)
```

```python
## Independent and dependent features
## Independent ----->> dataframe or 2-D array
x = df[['Weight (kg)']]
x
```

```python
np.array(x).shape
```

```python
## dependent ------>>  series or 1-D array
y = df['Height (cm)']
y
```

```python
np.array(y).shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.25, random_state= 45)
```

```python
##standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
```

```python
x_train
```

```python
x_test = scaler.transform(x_test)
```

```python
x_test
```

```python
## Apply simp;e linear regression
from sklearn.linear_model import LinearRegression

regression = LinearRegression(n_jobs=-1)
```

```python
regression.fit(x_train, y_train)
```

```python
print("gradient: ", regression.coef_)
print("intercept: ", regression.intercept_)
```

```python
#plot the train data best fit line
plt.scatter(x_train, y_train)
plt.plot(x_train, regression.predict(x_train))
```
